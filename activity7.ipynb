```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import folium
from folium.plugins import HeatMap
```

```python
df = pd.read_csv('Global_terrorist.csv', encoding='ISO-8859-1')
```

```python
df
```

<H1>Filtering Data</H1>

```python
filter1 = df[df['region_txt'] == "South Asia"]
```

```python
filter2 = filter1[filter1['attacktype1_txt'] == 'Armed Assault']
```

```python
filter3 = filter2[filter2['nkill'] >= 5]
```

```python
    filter4 = filter3[filter3['iyear'] > 2010]
```

```python
print(filter4.head())
```

```python
SA_terror = filter4.groupby(['longitude', 'latitude']).size().reset_index(name='incident_count')
```

```python
print(SA_terror.head())
```

<h1>VISUALIZATION OF GEOSPATIAL USING FOLIUM</h1>

```python
heat_data = SA_terror[['latitude', 'longitude', 'incident_count']].values.tolist()
```

```python
map_sa = folium.Map(location=[20, 30], zoom_start=3)
```

```python
HeatMap(heat_data).add_to(map_sa)
```
